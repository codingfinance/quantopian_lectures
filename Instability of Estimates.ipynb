An estimate by definition has uncertainties. Sometimes professionals in finance ignore this uncertainty. This can lead to heavy losses. In this notebook we will try to get familiar with some of these uncertainties with estimates.

In finance and investing the data comes in time series. We can calculate a set of mean and standard deviation of this this data. But this estimate of mean and standard deviation will change as we incorporate more data into our analysis. Lets look at some examples.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
normal = np.random.normal(0,1,500)
```

```python
plt.hist([normal[:10], normal[10:100], 
          normal[100:250], 
          normal], 
         density=1, histtype='bar', stacked=True, bins = 30);
```

```python
print(np.std(pd.DataFrame(normal).sample(5).values))
print(np.std(pd.DataFrame(normal).sample(50).values))
print(np.std(pd.DataFrame(normal).sample(200).values))
print(np.std(pd.DataFrame(normal).sample(175).values))
```

From the above chart we can see that taking different samples of different size from the same data can give different answers for our standard deviation estimates.

### Non Normal distribution

Sometimes mean and the standard deviation can be highly uninformative. 

```python
# Generate Bimodal Distribution

def bimodal(n):

    X = np.zeros((n))

    for i in np.arange(n):
        if np.random.binomial(1,0.5) == 0:
            X[i] = np.random.normal(-5,1)
        else:
            X[i] = np.random.normal(5,1)
    return X
        
    

X = bimodal(1000)

# plotting the distribution
plt.figure(figsize=(10,8))
plt.hist(X, bins = 50)
plt.ylabel("Frequency")
plt.xlabel("Value")
print(f"The mean of the distribution is {np.mean(X)}")
print(f"The standard deviation of the distribution is {np.std(X)}");
```

In the above chart the mean is not informative of the underlying distributions. So our estimate of mean in this case can lead to bad results.

```python
from statsmodels.stats.stattools import jarque_bera

_,pvalue,_,_ = jarque_bera(X)

print(f"The p-value is {pvalue}")
print('-'*50)

if pvalue > 0.05:
    print('The distribution is likely normal')
else:
    print("The distribution is likely not normal")
print('-'*50)

print(f"At 5% interval, the p-value {pvalue} is lower than 0.05, we can \
conclude that the distribution is not normal")
```

### Sharpe Ratio

Another singal point estimate that gets used a lot in evaluating a investor is the Sharpe Ratio. Sharpe ratio measures the additional returns achieved for additional risk. It lets investors quickly compare one manager with another. But this singal point estimate can be sometimes deceiving. <br>

There are often periods when managers will have a low sharpe ration, indicating the periods when strategy did not produce great results. To observe this, we need to calculate the rolling sharpe ratio. 

Lets us consider the CGM Focus Fund. Throughout the early 2000s this fund produced great result. It also experienced periods with high volatility. Lets see how this fund performed in light of the volatility.

```python
import pandas_datareader as web

fund_ret = web.get_data_yahoo('CGMFX', '1999-1-1')['Adj Close'].pct_change()[1:]
```

```python
risk_free = web.get_data_fred('DGS3MO', '1999-1-1')
```

```python
risk_free.head()
```

Join both the data frames.

```python
df = risk_free.join(fund_ret).dropna()
```

```python
df.head()
```

Chnage the column names

```python
df.columns = ['tbill', 'fund']
```

Convert the tbill to daily returns

```python
df['tbill'] = df['tbill']/100/365
```

```python
df.head()
```

```python
def sharpe_ratio(asset, riskfree):
    
    return np.mean(asset - riskfree) / np.std(asset - riskfree)

running_sharpe = [sharpe_ratio(df['fund'][i-90:i], df['tbill'][i-90:i]) for i in range(90,len(df))]
```

```python
sr_df = pd.DataFrame(data=running_sharpe, index = df.index[90:],
             columns=(['sharpe_Ratio']))
sr_df.plot(figsize=(12,8))
plt.xlabel('Date')
plt.ylabel('Sharpe Ratio')
plt.title("90 Day Rolling Sharpe Ratio\n for CGM Focus Fund");
```

The Sharpe ratio looks highly volatile and thus highly unreliable as a single point estimate.

We can look at the mean Sharpe Ratios and also the +/- 1 standard deviations of the Sharpe Ratio.

```python
mean_rs = np.mean(sr_df[:-100])
std_rs = np.std(sr_df[:-100])

# Plot running Sharpe ratio
_, ax2 = plt.subplots(figsize=(10,8))

ax2.plot(sr_df)

# Plot its mean and the +/- 1 standard deviation lines
ax2.axhline(mean_rs[0], color = 'black')
ax2.axhline(mean_rs[0] + std_rs[0], linestyle='--', color = 'red')
ax2.axhline(mean_rs[0] - std_rs[0], linestyle='--', color = 'red')

# Indicate where we computed the mean and standard deviations
# Everything after this is 'out of sample' which we are comparing with the estimated mean and std
# ax2.axvline(len(returns) - 100, color='pink');
plt.xlabel('Date')
plt.ylabel('Sharpe Ratio')
plt.legend(['Sharpe Ratio', 'Mean', '+/- 1 Standard Deviation'])

print('Mean of running Sharpe ratio:', mean_rs)
print('std of running Sharpe ratio:', std_rs)
```
