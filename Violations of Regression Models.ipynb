### Heteroskedasticity

When the variance of the error is not constant across observations, our data may display heteroskedasticity. 

Lets plot the data to visually observe what it means to have heteroskedasticity.

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels import regression
from statsmodels.stats.stattools import jarque_bera
import statsmodels.api as sm
plt.style.use('fivethirtyeight')
```

First lets observe Normal distribution and use the Jarque Bera test to confirm the normality.

```python
residual_normal = np.random.normal(0,1,1000)
_,pvalue_normal,_,_ = jarque_bera(residual_normal)
plt.hist(residual_normal,ec='b')
print(f"pvalue-{pvalue_normal} for normal distribution is > 0.05 significan level")
```

```python
residual_poisson = np.random.poisson(size=1000)
_,pvalue_poisson,_,_ = jarque_bera(residual_poisson)
plt.hist(residual_poisson,ec='b')

print(f"pvalue-{pvalue_poisson} for poisson distribution is < 0.05 significance level")
```

```python
x = np.arange(100)
y1 = x + 3 * np.random.randn(100)

sl1 = regression.linear_model.OLS(y1,sm.add_constant(x)).fit()
fit1 = sl1.params[0] + sl1.params[1] * x

plt.scatter(x,y1, alpha = 0.5, color = 'red')
plt.plot(x,fit1)
plt.title('Homoskedastic errors');
plt.legend(['Predicted', 'Observed'])
plt.xlabel('X')
plt.ylabel('Y');
```

```python
y2 = x * (1 + 0.5 * np.random.randn(100))

sl2 = regression.linear_model.OLS(y2,sm.add_constant(x)).fit()
fit2 = sl2.params[0] + sl2.params[1] * x

plt.scatter(x,y2, alpha = 0.5, color = 'red')
plt.plot(x,fit2)
plt.title('Heteroskedastic errors');
plt.legend(['Predicted', 'Observed'])
plt.xlabel('X')
plt.ylabel('Y');
```

### Testing for heteroskedasticity

We can test for Heteroskedasticity using the Breusch-Pagan Lagrange Multiplier test for heteroscedasticity from the statsmodels. The null hypothesis is that the data is Homoscedastic.

```python
residuals1 = y1-fit1
residuals2 = y2-fit2
```

Lets plot just the residuals.

```python
plt.figure(figsize=(10,8))
plt.scatter(range(len(residuals1)),residuals1, label = "Residual 1")
plt.scatter(range(len(residuals2)),residuals2, label = "Residual 2")
plt.legend();
```

```python
_, jb_pvalue1, _, _ = jarque_bera(residuals1)
_, jb_pvalue2, _, _ = jarque_bera(residuals2)

print("p-value for residuals1 ", jb_pvalue1)
print("p-value for residuals2 ", jb_pvalue2)
```

```python
x_with_constant = sm.add_constant(x)

_, pvalue1, _, _ = sm.stats.diagnostic.het_breuschpagan(residuals1, x_with_constant)
_, pvalue2, _, _ = sm.stats.diagnostic.het_breuschpagan(residuals2, x_with_constant)

print("p-value for residuals1 being heteroskedastic is ", pvalue1)
print("p-value for residuals2 being heteroskedastic is ", pvalue2)
```
